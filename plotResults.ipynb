```python
# importing required libraries
import pandas as pd
import matplotlib.pyplot as plt
import math
import numpy as np
```

```python
ga_df = pd.read_csv("genetic_results2.csv")
ga_df.head()
```

```python
ma_df = pd.read_csv("memetic_results.csv")
ma_df.head()
```

```python
# Comparing ga and ma for each layout

minimax_ga_df = (ga_df.loc[ga_df['layout'] == "minimaxClassic"]).reset_index()
minimax_ma_df = (ma_df.loc[ma_df['layout'] == "minimaxClassic"]).reset_index()

plt.plot(minimax_ga_df['bests'], label="genetic")
plt.plot(minimax_ma_df['bests'], label="memetic")
plt.legend()
```

```python
small_ga_df = (ga_df.loc[ga_df['layout'] == "smallClassic"]).reset_index()
small_ma_df = (ma_df.loc[ma_df['layout'] == "smallClassic"]).reset_index()
plt.plot(small_ga_df['bests'], label="genetic")
plt.plot(small_ma_df['bests'], label="memetic")

plt.legend()
```

```python
medium_ga_df = (ga_df.loc[ga_df['layout'] == "mediumClassic"]).reset_index()
medium_ma_df = (ma_df.loc[ma_df['layout'] == "mediumClassic"]).reset_index()
plt.plot(medium_ga_df['bests'], label="genetic")
plt.plot(medium_ma_df['bests'], label="memetic")
plt.legend()
```

```python
plt.clf()
original_ga_df = (ga_df.loc[ga_df['layout'] == "originalClassic"]).reset_index()
original_ma_df = (ma_df.loc[ma_df['layout'] == "originalClassic"]).reset_index()
plt.plot(original_ga_df['bests'], label="genetic")
plt.plot(original_ma_df['bests'], label="memetic")
plt.legend()
```

```python
capsule_ga_df = (ga_df.loc[ga_df['layout'] == "capsuleClassic"]).reset_index()
capsule_ma_df = (ma_df.loc[ma_df['layout'] == "capsuleClassic"]).reset_index()
plt.plot(capsule_ga_df['bests'], label="genetic")
plt.plot(capsule_ma_df['bests'], label="memetic")

plt.legend()
```

```python
# Comparing layouts
plt.plot(capsule_ga_df['averages'], label="capsule")
plt.plot(small_ga_df['averages'], label="small")
plt.plot(medium_ga_df['averages'], label="medium")
plt.plot(minimax_ga_df['averages'], label="minimax")
plt.plot(original_ga_df['averages'], label="original")

plt.legend()
```

```python
# Comparing layouts
plt.plot(capsule_ma_df['bests'], label="capsule")
plt.plot(small_ma_df['bests'], label="small")
plt.plot(medium_ma_df['bests'], label="medium")
plt.plot(minimax_ma_df['bests'], label="minimax")
plt.plot(original_ma_df['bests'], label="original")

plt.legend()
```
